```python
# Merge train data with store information
merged_data = pd.merge(train, stores, on='store_nbr', how='left')

# Merge with oil data
merged_data = pd.merge(merged_data, oil, on='date', how='left')

# Merge with transactions data
merged_data = pd.merge(merged_data, transactions, on=['date', 'store_nbr'], how='left')
```

```python
# For holidays, we need to handle different locale levels (National, Regional, Local)
# First, create national holidays
national_holidays = holidays[holidays['locale'] == 'National'][['date', 'type', 'description', 'is_transferred', 'is_effective_holiday']]
national_holidays = national_holidays.rename(columns={'type': 'national_holiday_type', 
                                                    'description': 'national_holiday_desc'})
```

```python
# Merge national holidays
merged_data = pd.merge(merged_data, national_holidays, on='date', how='left')
```

```python
# Create regional holidays (merge by date and state)
regional_holidays = holidays[holidays['locale'] == 'Regional'][['date', 'locale_name', 'type', 'description', 'is_transferred', 'is_effective_holiday']]
regional_holidays = regional_holidays.rename(columns={'locale_name': 'state',
                                                    'type': 'regional_holiday_type',
                                                    'description': 'regional_holiday_desc'})
```

```python
# Merge regional holidays
merged_data = pd.merge(merged_data, regional_holidays, on=['date', 'state'], how='left')
```

```python
# Create local holidays (merge by date and city)
local_holidays = holidays[holidays['locale'] == 'Local'][['date', 'locale_name', 'type', 'description', 'is_transferred', 'is_effective_holiday']]
local_holidays = local_holidays.rename(columns={'locale_name': 'city',
                                              'type': 'local_holiday_type',
                                              'description': 'local_holiday_desc'})
```

```python
# Merge local holidays
merged_data = pd.merge(merged_data, local_holidays, on=['date', 'city'], how='left')
```

```python
# Create consolidated holiday flags
merged_data['is_holiday'] = (~merged_data['national_holiday_type'].isna()) | \
                           (~merged_data['regional_holiday_type'].isna()) | \
                           (~merged_data['local_holiday_type'].isna())

merged_data['holiday_type'] = merged_data['national_holiday_type'].fillna(
    merged_data['regional_holiday_type'].fillna(
        merged_data['local_holiday_type']
    )
)
```
